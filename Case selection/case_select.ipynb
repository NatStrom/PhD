```python
import pandas as pd
import os
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
os.getcwd()
```

```python
dta = pd.read_csv('sanction_cases.csv')
print(dta.columns)
```

```python
dta_cases = dta.filter(('cname','year','sanction','success','conflict','one_sided_violence','best_fatality_estimate','ht_regtype',
                         'v2x_polyarchy', 'v2x_civil_liberties_index','impositionEU','impositionUS','impositionUN','impositionEU_type','impositionEU_econ'
                        'impositionEU_target','impositionUS_type','impositionUS_econ','impositionUS_target','impositionUN_type','impositionUN_target','impositionUN_econ'))
dta_cases.head(20)
```

```python
dta_cases = dta_cases.dropna(subset='sanction')
dta_cases
```

```python
df = dta_cases.copy()
```

```python
# Ensure "impositionEU_type" is treated as a list of values
df["impositionEU_type"] = df["impositionEU_type"].astype(str).str.split(";")

# Explode the column so each value gets its own row
df_exploded = df.explode("impositionEU_type")

# One-hot encode the values with 'EU' as prefix
dummies = pd.get_dummies(df_exploded["impositionEU_type"], prefix="EU")

# Merge back by grouping on the original index and taking the max (boolean OR)
df = df.drop(columns=["impositionEU_type"]).join(dummies.groupby(df_exploded.index).max())

# Display the transformed DataFrame columns
print(df.columns)
```

```python
# Ensure "impositionEU_type" is treated as a list of values
df["impositionUS_type"] = df["impositionUS_type"].astype(str).str.split(";")

# Explode the column so each value gets its own row
df_exploded = df.explode("impositionUS_type")

# One-hot encode the values with 'EU' as prefix
dummies = pd.get_dummies(df_exploded["impositionUS_type"], prefix="US")

# Merge back by grouping on the original index and taking the max (boolean OR)
df = df.drop(columns=["impositionUS_type"]).join(dummies.groupby(df_exploded.index).max())

# Display the transformed DataFrame columns
print(df.columns)
```

```python
# Ensure "impositionEU_type" is treated as a list of values
df["impositionUN_type"] = df["impositionUN_type"].astype(str).str.split(";")

# Explode the column so each value gets its own row
df_exploded = df.explode("impositionUN_type")

# One-hot encode the values with 'EU' as prefix
dummies = pd.get_dummies(df_exploded["impositionUN_type"], prefix="UN")

# Merge back by grouping on the original index and taking the max (boolean OR)
df = df.drop(columns=["impositionUN_type"]).join(dummies.groupby(df_exploded.index).max())

# Display the transformed DataFrame columns
print(df.columns)
```

```python
df_filter = df[(df['EU_7'] == True) | (df['EU_7b'] == True) | (df['EU_9'] == True)| (df['US_7'] == True)| (df['US_7b'] == True)| (df['US_9'] == True)| (df['UN_9'] == True)| (df['UN_7'] == True)| (df['UN_7b'] == True)]
df_filter
```

```python
#create a plot grouping it based on the polyarchy and civil liberties index
plt.figure(figsize=(20, 10))
sns.histplot(data=df_filter, x="cname", y="year", hue="impositionUS_target", palette="tab10", edgecolor="black")
plt.show()
```

```python
#create a plot grouping it based on the polyarchy and civil liberties index
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_filter, x="v2x_polyarchy", y="v2x_civil_liberties_index", hue="cname", palette="tab10", edgecolor="black")
plt.title("Polyarchy vs Civil Liberties Index of Sanctioned Countries with at least one type of targeted sanctions imposed")
plt.xlabel("Polyarchy Index (Higher = More Democratic)")
plt.ylabel("Civil Liberties Index (Higher = Less Free)")
plt.grid(True)
plt.show()
```
